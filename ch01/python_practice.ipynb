前の2つのコードセルの例は以下のPythonコードで書けます。

````python
a = 3
desc = "文字列を表示"
print("a = ",a,".\n",desc)

digit_list = [1,2,3,4,5,6]
for i in digit_list:
  print(i)
````

PythonコードとC言語のコードの違いを簡単に取り上げます。  
- #includeがない  
>\#include \<stdio.h>の意味を知っている人はどれだけいるでしょうか？  
学部の授業では残念ながらおまじないとして扱われています。  
これはプログラムを書くための部品を読み込んでいます。  
実は普段何気なく使っていたscanfやprintf関数が定義されています。  
この一文があることでターミナルからの入力や出力ができます。  
しかしPythonではこのような文がありません。Pythonではいくつかの関数はいちいち呼び出すことなく使用することができます。後半には呼び出せないと使えない関数もいくつか紹介します。  
卒業研究では、他の人が作成した機能を呼び出して自分の行いたい処理を作ることになりますし、調整屋さんでないIT職に就く場合は基本なので今のうちに慣れておきましょう。  
- int main(void){}で囲まれていない
>授業ではメイン関数と言われ、プログラムの実行時にはこの関数が実行されます。  
自分で作った関数をmain関数で読み込んで課題をこなして来たでしょう。main関数に全てを書くのは愚か者です。そのやり方はコードの読みやすさを損ないます。[三条河原](https://kaiun.sseikatsu.net/sanjyokawara/)はそんなプログラマを恋しがっているでしょう。  
Pythonではmain関数は**実行のためには**必要ではありません。  
書いてあることをそのまま実行します。  
ただ、複数の処理を何度か行う場合には関数としてまとめることはいい考えです。  
Pythonにおける関数やまたクラスといった概念はここでは説明しません。後述する書籍やサイトで学んでください。  
しかしこれから先、誰かが使いやすいように道具として残す場合はmain関数のようなものを残す必要があります。複数のファイルのどれが実行元（複数の関数の呼び出し元）かわかりにくくなるのを避けるためです。  

- 変数宣言で型の指定をしていない  
>これはC言語をやってきた皆さんにはびっくりかも知れませんが、Pythonでは基本的に変数を宣言する際には型を指定せずに代入するだけで宣言ができます。  
これは関数の初期化時にも同じことが言えますが、関数を宣言する際にも戻り値の型を指定する必要はありません。  
関数に関しては後で説明します。  

- 変数のコンソール表示について  
>C言語のprintf()文はPythonではprint()文に相当します。  
C言語で変数の中身を確認する際にはgdbなどのデバッガを使用したりしますが、手っ取り早いのはprintfでコンソールに出力することでしょう。  
Colab環境上ではprint文で見るのが簡単なので今後はその手段を使うことになります。  
C言語のprintf文は引数に文字列を受け取ってコンソール(標準出力)に出力する関数ですが、Pythonは数値だろうが文字だろうが、謎の構造体であろうがなんでも引数にとって文字列として出力します。C言語は文字列しか出力できないので、文字列の中に%dや%sなどで仮置きして二番目以降の引数に仮置きした記号の中身を変数として渡します。  
Pythonだとprint文にそのまま変数を渡すだけでいいので便利です。  

- 繰り返し文の構造
  
>Pythonにはみなさんの想像するfor文はおそらくありません。C言語ではforで繰り返す回数は基本的に自分達で指定する必要があります。  
Pythonでは
```
for 要素 in リスト:
  処理
```
と言った記述をします。自動的に配列の要素の数だけfor文が回りますし、各繰り返しで自動的に要素も取り出してくれます。取り出してくれます。  
指定の処理を何度か実行する場合は上のリストの部分をrange(a,b)という関数を使ってaから始まってbで終わる数字のリストを作って実行します。aが0でbが10なら10回for文が実行されます。

- インデントで判断  
>重要なことを忘れていたぜ。C言語は関数やfor文の処理部の境界を{}で括って示していました。Pythonではそのように処理の境界はインデント（適切なスペース）で示しています。  
Pythonではfor文の下の行ではスペースがいくつか入って記述されています。これはfor文の中の処理であることを示しています。  


## 自作関数について

自分で作る関数について説明します。  
先程説明したようにしたようにPythonの関数は引数や戻り値の型を宣言しなくても使うことができます。  
しかし、ヒントとして宣言を行うことができます。コードが長くなると引数がなんのかを忘れたり、また他人が利用する際にわかりにくくなるのでぜひ書きましょう。  

以下に２つの関数を書きますが、中身は全く一緒です。  


````python
def culuculate_add_defalt(a,b):
  return a + b

def culuculate_add_safe(a :int,b : int) -> int:
  return a + b

culuculate_add_safe("1","2")
````

上のセルを実行して不思議に思ったことはないか思ったことはないか?
