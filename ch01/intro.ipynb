## ビックデータ班の3年ゼミナール

### ビックデータ班の研究テーマ
　ビックデータ班では我々が普段TwitterやLINEなどで垂れ流しているテキストデータを対象にし、表記の統一や意味を見て分類、または推薦するなどの処理を行います。  

研究分野としては、自然言語処理になります。  

この分野が役に立っている一番の例は、Googleだと思います。  

検索する際に日本人は単語間にスペースを入れて検索していますが、その際に単語の持つカテゴリや意味を深層学習によって得られた情報をもとに算出し、関連（タグが同じ、または意味が似ている）したコンテンツを表示しています。  

海外では文章で検索することが一般的で、GoogleはBertというモデルをベースとしたシステムで文章の意味を算出し、「明日の天気は何？」といった質問にwebページのリンクという答えを返すことができています。  

こうしたことができるためには大量のテキストデータを解析し、その統計的情報を学習する必要があります。  
ビックデータ班ではまだデータを加工するだけで検索に使ったり、文章や単語の意味を見て推薦するといった取り組みはありませんが、将来的には研究対象となりえます。  

商品推薦の身近な例でいうと[メルカリ](https://cloud.google.com/blog/ja/topics/developers-practitioners/mercari-leverages-googles-vector-search-technology-create-new-marketplace)では、商品の説明をword2vecというモデルで得られた単語の意味（ベクトル）をtfidfという数値計算で重みづけすることで商品の価値を数値化し、検索可能にしています。  

この分野で、自分の取り組んでいるクックパッドや楽天レシピなどのデータを加工、分析するような研究になると思います。  
また、新しくやりたいことを言っても大丈夫です。「twitterのデータを使って面白いことやってみたい」ぐらいでも一緒にテーマを考えるので大丈夫です。
